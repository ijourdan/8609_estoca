## Espacio Muestral.

Si se realiza un experimento, un resultado particular de ese experimento se lo denomina **muestra puntual**,  $\zeta$. El conjunto de todos los resultados obtenibles en un experimento es el **espacio muestral** $ S=\left\{\zeta _i\right\}$  con $i \in \mathbb{N}$. 

Como ejemplo, de **espacio muestral discreto**, planteamos el experimento de arrojar una moneda al aire tres veces y anotamos la secuencia que devuelve, cara: $C$ o seca: $S$. De tal manera, la muestra número 40 del experimento es $\zeta^1 _{40}=CSS$ siendo el espacio muestral:

$$S_1=\left\{CCC, CCS, CSC, CSS, SCC, SCS, SSC, SSS\right\}$$ 

Un ejemplo de **espacio muestral continuo**, y muy común en la industria de las luminarias, es determinar la vida útil de una lámpara. Para ello se encienden un conjunto de lámparas y se registran los tiempos en que demoran las lámparas hasta registrar una falla. Para este experimento, la k-ésima muestra puede ser $ \zeta_2 ^k = 475.234 seg $, unos 11 meses de duración.

El espacio muestral está formado por los tiempos de vida $t_2 ^k$ que fueron medidos para cada lámpara. El tiempo, abordado de esta manera, es una variable continua, y por otro lado, no se conoce un tiempo máximo de vida de tales lámparas, por lo tanto, puede considerarse como espacio muestral:

$$S_2=\left[0,\infty\right)$$ 

## Eventos

Un evento es un subconjunto del espacio muestral $S$. Se puede definir al evento $A_1$ como *vida útil menor a una hora para las lámparas*, en otros términos:

$$A_1=\left\{t\in S ~/~ t\leq 3600 seg\right\}$$ 

El evento así definido, se puede observar mediante el estudio de un conjunto de muestras adquiridas en un experimento apropiado, en ese sentido, podemos decir que "sucedió tal evento porque así lo reflejó una muestra". 

Entonces, quién importa es el evento, y no el suceso particular de una muestra, la cuál es válida únicamente en el instante que se adquirió. 

Así como podemos plantear el evento $A_1$, también es posible plantear el evento $A_2=\{t\in S ~/~ t=12750seg\}$ o bien, las lámparas cuya vida útil es exactamente 12750 segundos. También aquí hablamos de un evento, el cuál se expresa cuando una lámpara se quema exactamente a los 12750 segundos.

Se define el Conjunto Potencia $ 2^S $, y a partir de éste define un conjunto numerable de subconjuntos de $S$ el cual se estructura en una $\sigma$-álgebra $\mathcal{F} \subset 2^S$. Es decir, cumple: 

* El conjunto vacío $\emptyset$ está en $\mathcal{F}$ ($\emptyset \subset \mathcal{F}$)
* $\mathcal{F}$ es cerrado en el complemento, es decir, $A \subset \mathcal{F} \Rightarrow \overline{A} \subset \mathcal{F}$
    * Siendo $\overline{\emptyset} = S$, entonces, $S\subset \mathcal{F}$
    * $\mathcal{F}$ es cerrado para la intersección. Si $A_i \subset \mathcal{F}$ con $i=1,2,3,...$, entonces $\displaystyle\cap_{i=1}^{\infty} A_i \subset \mathcal{F}$
    

Queda estructurado en $\mathcal{F}$ los eventos que buscamos medir. Definiendo una medida $m()$ sobre $\mathcal{F}$, podemos entonces medir los eventos, supongamos $A_j, A_k \in \mathcal{F}$, siendo tales medidas $m(A_j)$ y $m(A_k)$. Si nos preguntamos sobre dos nuevos eventos, $A_l=\left\{t\in S ~/~ t\in A_j y t\in A_k\right\}$,  es decir $A_l=A_j \cap A_k$, el cual está en $\mathcal{F}$ y por lo tanto es posible medirlo con $m()$

## Medida de Probabilidad.

Se define una $\sigma-$álgebra $\mathcal{F}$. Una medida de probabilidad $P$ sobre $\mathcal{F}$ es una medida que asigna a cada evento $A\in\mathcal{F}$ un número $P(A)$ satisfaciendo los axiomas:

1. $\forall A \in \mathcal{F}, P(A) \geq 0$
2. Siendo S el espacio muestral, entonces $P(S)=1$
3. Siendo $A_1,A_2,A_3,.... \in \mathcal{F}$ una sucesión de **eventos disjuntos**, $\forall i \neq j, A_i \cap A_j =\emptyset$, entonces: 

$$ P\left [ \displaystyle\bigcup_{i=1}^{\infty} A_i \right ] = \sum_{i=1}^{\infty} P\left [ A_k \right ] $$

Si $S$ fuera un espacio muestral finito el tercer axioma se reduce a una unión finita, y por lo tanto, una suma finita.

Definida la medida de probabilidad, se está en condiciones de estructurar un **espacio de probabilidad** el cuál se nota como $(S,\mathcal{F},P)$, siendo $S$ el espacio muestral, $\mathcal{F}$ la $\sigma$-álgebra y $P$ la medida de probabilidad.

### Complemento.
 
Siendo $A\in\mathcal{F}$ un evento con probabilidad $P(A)$ , su complemento $\overline A$ tiene probabilidad $P(\overline{A}) = 1-P(A)$.

La demostración es simple. $A\cap \overline A = \emptyset$ y además $A \cup \overline A =S$, luego $P(A)+P( \overline A) =1$, quedando demostrado.

* Siendo $\overline S = \emptyset$ entonces, $P(\emptyset)=0$

## Probabilidad condicional

Se definen dos eventos $A,B \in \mathcal{F}$, se conoce que ocurre $B$, es decir, existe una condición que impone la ocurrencia del evento $B$. Bajo esas condiciones interesa determinar la probabilidad de que ocurra el evento $A$.
Estamos buscando la probabilidad $A$ condicionada por $B$: $P\left ( ^{A}/_{B} \right )$.

En forma intuitiva, si se condiciona a la expresión de $B$, la expresión de $A$ estará dada por la expresión de $A\cap B$, luego podemos estimar la probabilidad condicional:

$$\lim_{n \to \infty} \frac{n_{A\cap B}}{n_B} \rightarrow  P\left ( ^{A}/_{B} \right )$$

siendo $n_{A\cap B}$ y $n_B$ la cantidad de veces que se expresan los eventos $A\cap B$ y $B$ respectivamente. Sea $n$ la cantidad total de experimentos, podemos escribir:


$$\lim_{n \to \infty} \frac{\frac{n_{A\cap B}}{n}}{\frac{n_B}{n}} \rightarrow  \frac{P\left ( {A}\cap{B} \right )}{P\left ({B} \right )}= P\left ( ^{A}/_{B} \right )$$

Luego, $P\left ( {A}\cap{B} \right ) = P\left ({B} \right ) ~ P\left ( ^{A}/_{B} \right )$, o bien, $P\left ( {A}\cap{B} \right ) = P\left ({A} \right ) ~ P\left ( ^{B}/_{A} \right )$.


### Regla de Bayes
Se tiene una partición $B_1, B_2, B_3, ..., B_n$ del espacio muestral $S$, y ocurre un evento $A$, luego
$$ P\left ( ^{B_j}/_{A} \right ) \sum_{i=1} ^n P\left ( ^{A}/_{B_i} \right ) P(B_i)= P\left ( ^{B_j}/_{A} \right ) P(A)=P\left ( ^{A}/_{B_j} \right ) P(B_j)$$

### Eventos Independientes.

Se definen dos eventos $A$ y $B$, se busca determinar la probabilidad de ocurrencia de $A$ condicionado a $B$. Si el evento $B$ no condiciona la ocurrencia del evento $A$, entonces decimos que son independientes, y por lo tanto:

$$P\left ( ^{A}/_{B} \right )=P(A)$$

o en forma equivalente,

$$P\left ( ^{A}/_{B} \right )= \frac{P\left ( {A}\cap{B} \right )}{P\left ({B} \right )}=P(A)$$

por lo tanto, si dos eventos, $A$ y $B$ , son independientes, la probabilidad de que se expresen ambos eventos es igual al producto de las probabilidades de expresión de los eventos:

$$P\left ( {A}\cap{B} \right )=P(B)~P(A)$$

## Variable Aleatoria.
Una variable aleatoria $X$ es una función que asigna un número real $X(\zeta)$, siendo $\zeta \in S$, una muestra del espacio muestral.

Entonces, una variable aleatoria no es más que un mapeo $X:S \rightarrow \mathbb{R}$ (o $\mathbb{C}$)

Se definió previamente al espacio muestral $S_2=\left[0,\infty\right)$, donde cada elemento $t \in S$ es un experimento que indica cuántos segundos funcionó una lámpara. Definimos la **variable aleatoria** $X(t)=t$. Luego, el evento  

$$A_1=\left\{t\in S ~\big /~ t\leq 3600 seg\right\}$$ 

puede ser referido mediante el mapeo como $X\leq3600$, y por lo tanto:

$$P(A_1)=P(X\leq3600)$$


### Variable aleatoria discreta.

Una variable aleatoria discreta $X$ se define como una variable que toma valores discretos a partir de un espacio muestral contable $S=\{s_1,s_2,s_3, ..... \}$. Siendo $A_k$ un evento del espacio muestral $S$, se lo puede definir como 

$$ A_k=\left \{ \zeta \in S \big / X(\zeta)=x_k \right \}$$

luego $P(A_k) = P(X=x_k)$. 

Se puede definir una **Función masa de probabilidad**, la cual es discreta y tiene la forma 

$$P_X(x)=P(X=x)$$

*Notar la diferencia en la notación, $P_X$ es la función masas de probabilidad, la cual tiene una definición de acuerdo al mapeo impuesto por la variable aleatoria. Por otra parte, $P$ es la medida de probabilidad sobre los eventos del espacio muestral.*

### Función distribución acumulativa de probabilidad.

Sea una variable aleatoria $X$, la **función de distribución acumulada** de define como la probabilidad del evento $A=\{a\in S \big / X(a) \leq x \}$:

$$ F_X (x)=P(A)=P\{X\leq x\}$$

#### Propiedades:


* $0\leq F_X (x) \leq 1$
* $\lim_{x \rightarrow \infty} F_X (x)=1$
* $\lim_{x \rightarrow -\infty} F_X (x)=0$
* Siendo $a \leq b$ con $a,b \in \mathbb{R}$ entonces $F_X(a) \leq F_X (b)$
* $F_X (x)$ es continua por derecha
* Se definen $a,b \in \mathbb(R)$, entonces $P(a< X \leq b)= F_X(b)-F_X(a)$
* $P(X>x)= 1- F_X(x)$

### Función densidad de probabilidad

Si existe, la **función densidad de probabilidad** de una variable aleatoria $X$, se define como:


$$f_X (x)= \frac {dF_X (x)}{dx}$$

donde, además:

$$ F_X(x) = \int _{-\infty} ^x f_X (x) ~ dx $$

#### Propiedades

* $f_X (x) \geq 0$

* $\int_{-\infty} ^\infty f_X(x)~ dx = 1$

* $P (a\leq X \leq b)= \int_a ^b f_X(x)~ dx$

### Algunos ejemplos

### Variable aleatoria exponencial

Una variable aleatoria es exponencial si su función de distribución acumulada tiene la forma:

$$ F_X (x) = \left \{ {\begin{eqnarray}
1- e^{-\lambda x} & & x \geq 0 \\
0 & & x<0\\
\end{eqnarray}} \right. $$

y su densidad de probabilidad es :

$$ f_X (x) = \left \{ {\begin{eqnarray}
\lambda e^{-\lambda x} & & x \geq 0 \\
0 & & x<0\\
\end{eqnarray}} \right. $$



```python
x=np.arange(-1,1,.01)
lamb=[0.1, 0.2, 0.9,1,2,4,8,10]
y=pd.DataFrame(index=x,columns=lamb,dtype='float')
for i in lamb:
    y.loc[x<0,i]=0
    y.loc[x>=0,i]=i*np.exp(-i*x[x>=0])
    
fig=plt.figure(figsize=(12,8))
ax=fig.add_subplot(1,1,1)
y.plot(ax=ax,lw=4,color=current_palette)
ax.grid(True)
ax.set_title('Función densidad de probabilidad exponencial: '+ r'$f_{X} (x)$')
ax.set_xlabel('x')


```

### Variable aleatoria Uniforme

Una variable aleatoria uniforme posee la función de distribución acumulada:

$$ F_X (x) = \left \{ {\begin{eqnarray}
0 &~& x < a \\
\frac {x-a}{b-a} &~& a \leq x \leq b \\
1 &~& x>b\\
\end{eqnarray}} \right. $$

y su dendidad de probabilidad es :

$$ f_X (x) = \left \{ {\begin{eqnarray}
\frac {1}{b-a} &~&  a \leq x \leq b \\
0 &~& x<a ~y~ x>b\\
\end{eqnarray}} \right. $$


```python
N=1e5
#parámetros:
b=3
a=2
#generamos nros aleatorios:
x=np.random.rand(int(N))
print('X:Unif(0,1)  max(x)='+str(x.max())+'   min(x)='+str(x.min()))

x=(b-a)*x+a
print('X:Unif('+str(a)+','+str(b)+')  max(x)='+str(x.max())+'   min(x)='+str(x.min()))
```

```python
fig=plt.figure(figsize=(12,8))
h1=plt.hist(x,bins=80,range=(0,4),density=True,color='#599AD3')
plt.grid()
plt.title('Histograma de una muestra de números aleatorios '+r'$X \in Unif(a,b)$')
```

```python
fig=plt.figure(figsize=(12,8))
h1=plt.hist(x,bins=80,range=(0,4),density=True,color='#599AD3',cumulative=True)
plt.grid()
plt.title('Histograma acumulado de una muestra de números aleatorios '+r'$X \in Unif(a,b)$')
```

## Cambio de variables


Sea $X$ una variable aleatoria,y se define el evento $A_X =\{ x\in \mathbb{R} \big / x=X(\zeta), ~ con~ \zeta \in S_0 ~ y ~ S_0 \in \mathcal{F} \} $. 

Se define también a $Y$ una variable aleatoria para la cual existe una función $g(.)$ tal que $Y=g(X)$. 

De acuerdo a las definiciones anteriores, podemos proponer que el evento $A_X$ es **mapeado** en la variable aleatoria $Y$ como el evento $A_Y$:

$$A_Y =\{ y\in \mathbb{R} \big / y=g \circ X(\zeta), ~ con~ \zeta \in S_0 ~ y ~ S_0 \in \mathcal{F} \} $$

Una pregunta puede ser ¿Cúal es el valor de $P(A_Y)$?

$$P(A_Y)= P(S_0) = P(A_X)$$

Entonces, los eventos $A_X$ y $A_Y$ poseen la misma probabilidad, es más, desde el punto de vista del espacio de probabilidad, son **representaciones** de un mismo evento $S_0$, $A_X$ es la representación del evento $S_0$ mediante la v.a. $X$, y $A_Y$ es la representación del mismo evento $S_0$, empleando la v.a. $g\circ X$.

### Cambio de variable y Distribución acumulada de Probabilidad.

Se conoce la variable aleatoria $X$, y se definió la función $g(.)$, de manera tal que $Y=g(X)$, por ejemplo, podemos suponer $y=g(x)=2x^2-2$, según se muestra en la siguiente figura:

![alt text](g_no_biyectiva.png "El código para graficar la figura se encuentra al final del texto.")

Se define el evento $Y_1=[y_a\leq y \leq y_b]$ , que para el caso del ejemplo, $y_a=2$ e $y_b=3$. La consigna puede ser, cuál es la probabilidad de que se exprese el evento $Y_1$, es decir $P(Y_1)$

Teniendo en cuenta que $Y=g(X)$, podemos observar en la gráfica que el evento que es mapeado en $Y_a$ puede representarse en $X$ como 

$$ X_a= X_1 \cup X_2 $$

de manera tal que 

$$Y_1= g(X_a)= g(X_1 \cup X_2) =g(X_1) \cup g(X_2)$$

donde $X_1=[-x_b,-x_a]$ y $X_2=[x_a,x_b]$

pero entonces, 

$$P(Y_1)= P(g(X_a))= P\left ( X_1 \cup X_2 \right )= P(X_1) + P(X_2) $$

por ser $X_1$ y $X_2$ dijuntos. 

Lo anterior puede reescribirse de otra manera. La función $g(.)$, no es biyectiva, y por lo tanto no existe una inversa, pero puede partirse en dos funciones biyectivas $g_1$ y $g_2$, de forma tal que puede expresarse:

$$ g(x) = \left \{ {\begin{eqnarray}
g_2(x) & & x \geq 0 \\
g_1(x) & & x<0\\
\end{eqnarray}} \right. $$

Por un lado, estas funciones $g_1$ y $g_2$ al ser biyectivas poseen inversa, y por otro, esta definición de $g(.)$ impone una partición del espacio en conjuntos disjuntos $(x \leq 0 ~ y ~ x<0)$, por lo tanto podemos escribir a la $P(Y_1)$ como: 

$$P(Y_1)= P(X_1) + P(X_2)=P\left (g_1^{-1}(Y_1) \right ) + P \left(g_2^{-1}(Y_1) \right ) $$



#### Distribución acumulada de probabilidad

Tomando como referencia la última ecuación, 

$$ P(Y_1)=F_Y(y=y_a)-F_y(y=y_b) $$

entonces podemos determinar la probabilidad conociendo la distribución acumulada de probabilidad. De igual forma,

$$F_Y(y)= P\left (g_1^{-1}([0,y]) \right ) + P \left(g_2^{-1}([0,y]) \right )= P\left (W_1 \right ) + P \left (W_2 \right )$$

donde 

$${\begin{eqnarray}
P(W_1)=F_X(x=0)-F_X(x=g_1^{-1}(y)) \\
P(W_2)=F_X(x=g_2^{-1}(y))-F_X(x=0)\\
\end{eqnarray}} $$

finalmente,

$$F_Y(y)= P\left (W_1 \right ) + P \left(W_2 \right )=F_X(x=g_2^{-1}(y))-F_X(x=g_1^{-1}(y))$$

* Observar que $F_Y(y)=F_X(-g_2^{-1}(y)\leq x \leq g_2^{-1}(y))$ en este caso.

En el ejemplo propuesto, $g_1^{-1}(y)=- \frac {1}{2} \sqrt{y+2}$  y  $g_2^{-1}(y)=\frac {1}{2} \sqrt{y+2}$

$$F_Y(y)= F_X \left (\frac {1}{2} \sqrt{y+2} \right )- F_X \left (-\frac {1}{2} \sqrt{y+2} \right )$$



#### En general.

Sea $Y=g(X)$ con $g:\mathbb{R} \rightarrow \mathbb{R}$, definido el evento $\left (y_a \leq y \leq y_a + \Delta y_a \right)$, entonces

$$\sum_{i=1}^{N} P\left( x^{(i)}\leq x \leq x^{(i)}+ \Delta x^{(i)} \right ) = P\left( y_a\leq y \leq y_a+ \Delta y_a \right ) $$

donde $x^{(i)}= g_i^{-1}(y_a)$, siendo $g_i$ la i-ésima función biyectiva tal que 

$$ g(x)=g_i(x) ~~ con ~~  ( x^{(i)}\leq x \leq x^{(i)}+ \Delta x^{(i)})$$

y se cumple

$$ ( x^{(i)}\leq x \leq x^{(i)}+ \Delta x^{(i)}) \cap ( x^{(j)}\leq x \leq x^{(j)}+ \Delta x^{(j)}) = \emptyset ~~ con ~~ i \neq j$$





### Cambio de variable. Función densidad de Probabilidad

Si existe la función densidad de probabilidad, entonces podemos aproximar $F_X(x+\Delta x)=f_X(x) \Delta x$, y considerando a $g(.)$ continua, existirá una aproximación $F_Y(y+\Delta y)=f_X(y) \Delta y$.  Podemos aproximar:

$$\sum_{i=1}^{N} F_X\left(  x^{(i)}+ \Delta x^{(i)} \right ) - F_X\left( x^{(i)}  \right ) =  F_Y \left(y+ \Delta y \right )- F_Y \left(y\right )  $$

por

$$\sum_{i=1}^{N} f_X\left(  x  \right ) \Delta x  =  f_Y \left(y  \right ) \Delta y  ~~ con~~  ( x^{(i)}\leq x \leq x^{(i)}+ \Delta x^{(i)}) $$

o bien

$$\sum_{i=1}^{N} \frac{f_X\left(  x  \right )} {\frac{\Delta y}{\Delta x}}  =  f_Y \left(y  \right )   ~~ con~~  ( x^{(i)}\leq x \leq x^{(i)}+ \Delta x^{(i)}) $$

para cada i-ésimo intervalo se tiene además

$$ {\frac{\Delta y}{\Delta x}} \rightarrow \frac{d g_i(x)}{dx} ~~ y ~~ g(x)=g_i(x) ~~ en ~~ ( x^{(i)}\leq x \leq x^{(i)}+ \Delta x^{(i)})$$ 

obteniendo finalmente la expresión de $f_Y(y)$:

$$f_Y \left(y  \right ) = \sum_{i=1}^{N} \left . \frac{f_X\left(  x  \right )} {\left | \frac{d g_i(x)}{dx}\right |}\right |_{x=g_i^{-1}(y)}      ~~ con~~  ( x^{(i)}\leq x \leq x^{(i)}+ \Delta x^{(i)}) $$


Retornando al ejemplo, 

$$\left | \frac{d g_i(x)}{dx}\right | = \left | 4 x \right |$$

luego

## Ejercicio 2

Se realizó el **Ejercicio 1**. $X\in \mathcal{Unif}[0,1]$, y $\{x_i\}$ son realizaciones de la v.a. $X$; el cambio de variables $g$:

$$y_i=g(x_i)=-\frac {1}{\lambda} ln(1-x_i) $$

genera $\{y_i\}$ realizaciones de una v.a. $Y\in exp(\lambda)$. Por otra parte, el cambio $g_1$:

$$z_i=g_1(y_i)= \sqrt {2 \alpha \lambda y_i} $$

genera $\{ z_i \}$ un conjunto de realizaciones $Rayleigh(\alpha)$.

Seguidamente, se generan muestras de las v.a. y se presentan sus histogramas.

```python
N=10000 #cantidad de muestras a generar
#Parametros:

lam=4 #parametro dist exponencial
a=3 #parametro dist Rayleigh

#gener Unif[0,1]
x=np.random.rand(N)

y=-1/lam*np.log(1-x)

z=np.sqrt(2*a*lam*y)

fig=plt.figure(figsize=(20,8))
ax=plt.subplot(1,3,1)
h1=plt.hist(x,bins=20,range=(-1,4),density=True,color='#599AD3')
plt.grid()
plt.title(r'$X \in Unif(0,1)$')

ax=plt.subplot(1,3,2)
h1=plt.hist(y,bins=20,range=(-1,5),density=True,color='#599AD3')
plt.grid()
plt.title(r'$Y \in Exp(\lambda)$')

ax=plt.subplot(1,3,3)
h1=plt.hist(z,bins=20,range=(-1,10),density=True,color='#599AD3')
plt.grid()
plt.title(r'$Z \in Rayleigh(a)$')
```

```python
print('Estimación de media y varianza de las v.a. Y y Z')
print('V.A. Exp(lambda='+str(lam)+'): m(Y)='+str(round(y.mean(),2))+ '     var(Y)='+str(round(y.var(),2)))
print('V.A. Ray(a='+str(a)+'):      m(Z)='+str(round(z.mean(),2))+ '     var(Y)='+str(round(z.var(),2)))
```

Variable aleatoria exponencia $Y$, con $\lambda=4$:

* Media teórica $\mu(Y)=\frac{1}{\lambda}$. Entonces $\mu(Y)=0,25$
* Varianza teórica $\sigma^2_Y = \frac{1}{\lambda^2}$. Entonces $\sigma^2_Y=0,0625$

Variable aleatoria Rayleig $Z$, con $\alpha=3$:

* Media teórica $\mu(Z)=\sqrt{\frac{\alpha \pi}{2}}$. Entonces $\mu(Z)=2,17$
