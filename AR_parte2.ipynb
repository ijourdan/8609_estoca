## Repaso:  Autocorrelación y Densidad Espectral de Potencias en sistemas LTI


<img src="img/lti.png" width="300">

La gráfica superior muestra un diagrama de un sistema LTI, donde $W(n)$, un proceso estocástico ESA, es la entrada a dicho sistema $h(n)$, siendo $x(n)$ el proceso aleatorio, también ESA, a la salida del sistema LTI. 

**Vamos a considerar siempre procesos ESA.**

Siendo $x(n) = h(n) \ast W(n)$ , con $( \ast) $ la operación de convolución. Definida la autocorrelación como $R_{xx}(m)=\mathbb{E}[x(n)~x(n+m)]$, se determina que la autocorrelación del proceso a la salida de del filtro $h(n)$ es:

$$ R_{xx} (m) = h(m) \ast R_{ww}(m) \ast h(-m)$$

La **Densidad Espectral de Frecuencias** $S_{xx}(\omega)$, de un proceso **ESA**, por el Teorema de Wiener - Khinchin, puede escribirse como la transformación por Fourier de $R_{xx}(m)$:

$$ S_{xx}(\omega) = \mathbb{F} \left( R_{xx}(m) \right ) ~~~con~|\omega|<\pi$$

por otra parte, $X(n)$ es la salida de un sistema LTI, luego podemos escribir:

$$S_{xx}(\omega) =  H(\omega) ~ S_{WW}(\omega) ~\overline{H(\omega)} = \left| ~ H(\omega) ~ \right| ^2 ~S_{WW}(\omega) $$




## Repaso: Proceso Autoregresivo

Se definió previamente al **Proceso autoregresivo de orden p (AR-p)** como:

$$ X(n) = \alpha_1~X(n-1) + \alpha_2~X(n-2) ~ + ~...~ +~\alpha_p~X(n-p) + W(n) = \sum_{i=1} ^p \alpha_i ~ X(n-i) + W(n) $$

donde $W(n)$ es ruido blanco con potencia $\sigma_W^2$. También se determinó que la autocorrelación del proceso AR responde a la ecuación recurrente que define el proceso:

$$ R_X(m) = \alpha_1~ R_X(m-1) + \alpha_2~R_X(m-2) ~ + ~...~ +~\alpha_p~R_X(m-p) = \sum_{i=1} ^p \alpha_i ~ R_X(n-i) $$


### Procesos AR-1

Se estudió el proceso AR-1, definido por la ecuación:

$$ X(n) = \alpha~X(n-1) + W(n) $$

con W(n) ruido blanco de varianza $\sigma_W^2$. El proceso está definido si el parámetro $\alpha$ cumple $-1<\alpha <1$. Finalmente, también se determinó que la autocorrelación del proceso AR-1 tiene la forma:

$$ R_X(m) = \sigma^2_X ~ \alpha^{|m|}$$

donde $\sigma_X^2 = {1\over {1-\alpha^2}}~\sigma_W^2$


### Procesos particulares - AR-2

Si es un autoregresivo de segundo orden (AR-2), el proceso tiene la forma:

$$ X(n) = \alpha_1~X(n-1) + \alpha_2~X(n-2) + W(n)  $$ 


con W(n) ruido blanco de varianza $\sigma_W^2$. También se analizó su autocorrelación: 


$$ R_X(m) = \alpha_1~ R_X(m-1) + \alpha_2~R_X(m-2)$$

resolviendo,

$$ R_{XX}(m)= A_1 ~ \lambda_1^{|m|} + A_2 ~ \lambda_2^{|m|}  ~~~~si ~~ \lambda_1 \neq \lambda_2$$

$$ R_{XX}(m)= \left( B_1  + B_2 ~|m| \right )~ \lambda^{|m|}  ~~~~si ~~ \lambda_1 = \lambda_2$$

siendo 

$$\lambda_{1,2}= {1 \over 2} ~ \left ( \alpha_1 \pm \sqrt{ \alpha_1 ^2 + 4\alpha_2} \right )$$

y si se cumple que $ \alpha_1 ^2 + 4\alpha_2 < 0$, entonces los $\lambda_i$ son complejos conjugados , transfiriendo un comportamiento oscilatorio amortiguado a $R_{XX}(m)$. Siendo $\lambda_{1,2}=|\lambda|~e^{jw_0}$ se tiene:

$$ R_{XX}(m)=  C  ~ |\lambda|^{|m|} ~ cos(w_0 ~ m) $$

Los coeficientes $A_1$, $A_2$, $B_1$, $B_2$ ó $C$  se determinan en base a las condiciones iniciales:

$$ R_{XX}(0)=\sigma_X^2$$ y $$R_{XX}(1) = R_{XX}(-1) = {{\alpha_1}\over{1-\alpha_2}} ~ \sigma_X^2$$

donde la varianza del proceso $X(n)$ se determinó como:

$$ \sigma_{X}^2 = {{1-\alpha_2} \over {(1+\alpha_2)(1-\alpha_1-\alpha_2)(1+\alpha_1-\alpha_2)}}~\sigma_W^2$$

y de esta última ecuación surgen condicionantes para los coeficientes $\left|\alpha_2\right|<1$ y  $|\alpha_1|<1-\alpha_2$

# Densidad Espectral de Potencia de un Proceso AR-2

Vamos a determinar la densidad espectral de potencia $S_{XX}(\omega)=\mathbb{F}\left (R_{XX}(m) \right )$ de un proceso $X(n)$ AR-2:

$$X(n)=\alpha_1 ~ X(n-1) + \alpha_2 ~ X(n-2) + W(n)$$

Teniendo en cuenta las bondades de una ecuación en diferencias, y empleando al **Transformada Z**, buscaremos otro camino, que implica determinar una función de transferencia:

$$ H(Z)={X(Z) \over W(Z)} = { 1 \over {1 - \alpha_1 ~ Z^{-1} - \alpha_2 ~ Z^{-2} }} $$

también podemos escribir esta transferencia como un producto de raíces del polinomio en el denominador.


$$ H(Z)= = { 1 \over {(1 - \lambda_1 ~ Z^{-1})~~(1 - \lambda_2 ~ Z^{-1}) }} $$ donde 


$$\lambda_{1,2}= {1 \over 2} ~ \left ( \alpha_1 \pm \sqrt{ \alpha_1 ^2 + 4\alpha_2} \right )$$

son los polos del filtro $H(Z)$, Entonces, 
- si los $\lambda_i$ son negativos, tendremos un filtro pasa-altos.
- si $\lambda_i$ fueran positivos, será pasabajos.
- si $\lambda_i$ fueran complejos conjugados, se logran cosas interesantes.

Entonces, sobre el círculo unitario, cuando $|Z|=1$, convergemos a transformada de Fourier de la transferencia: 

$$ H(\omega)= { 1 \over {1 - \alpha_1 ~ e^{-j\omega} - \alpha_2 ~ e^{-j2\omega} }} ~~~|\omega|<\pi$$


Conociendo la densidad de espectro de potencia del ruido $S_{WW}(\omega)=\sigma_W^2$, se tiene que la densidad  espectral de potencias del proceso es:

$$S_{xx}(\omega) =  H(\omega) ~ S_{WW}(\omega) ~\overline{H(\omega)} = \left| ~ H(\omega) ~ \right| ^2 ~S_{WW}(\omega) $$

operando, 

$$ S_{xx}(\omega)=  ~ { \sigma_W^2 \over {\left |1 - \alpha_1 ~ e^{-j\omega} - \alpha_2 ~ e^{-j2\omega} \right |^2}} ~~~|\omega|<\pi$$

### Resolución.
Se genera la función **SXX2(COEF,NFFT)** con la cual vamos a determinar la DEP, y sus polos.

```python
#Realizamos una función auxiliar
def SXX2(COEF,NFFT=1024):
    """
    Grafica la Densidad espectral de potencia para un AR-2 con ruido de varianza uno.
    CEOF: lista con los coeficientes del AR-2
    NFFT: cantidad de muestras (1024 por defecto)
    Retorna:PSD,pole
    PSD numpy.ndarray con DEP
    pole numpy.ndarray con polos
    """
    WW = np.arange(-3,3,6/NFFT)
    PSD = 1/np.abs(1-COEF[0]*np.exp(-1j*WW)-COEF[1]*np.exp(-1j*2*WW))**2 #DEP
    pole = np.roots([1,-COEF[0],-COEF[1]]) #determinamos polos
    
    fig = figure(figsize=(8,6))
    ax1 = fig.add_subplot(1,1,1)
    
    ax1.plot(linspace(-1, 1, len(PSD)), PSD)
    xlabel(r"Frecuencia Normalizada ($\times$ $\pi$ $rad/muestra$)")
    ylabel('DEP (Ampl/rad/sample)')
    t=title(r"Densidad Espectral de potencia del proceso AR2")
    
    return PSD,pole
```

```python
#COEF= [-.75, -.125] # ambos polos negativos
#COEF= [.75, -.125] # ambos polos positivos
#COEF= [-.4, -.04]   # ambos polos iguales
COEF= [.9, -.8]    # polos complejos conjugados
#COEF= [0.6, -.25]  # otros polos complejos conjugados
#COEF= [0, 0.25]    # polos reales de distinto signo

PSD,poles=SXX2(COEF)
print('Polos:')
print(poles)
print('V_absoluto')
print(abs(poles))
print('Ángulos (/pi)')
print(angle(poles)/np.pi)
```

```python
#Determinamos 
PSD2=np.roll(PSD,512)
fig = figure(figsize=(8,6))
plt.stem(np.real(np.fft.ifft(PSD2))[0:20])
xlabel(r" muestras $m \geq 0$ ")
ylabel('Ampl')
t=title(r"Autocorrelación $R_{XX}(m)$ del proceso AR2")


```

# Densidad Espectral de Potencia de un Proceso AR-p

No es difícil extender al caso general, siendo $X(n)$ un proceso AR-p:

$$ X(n) = \alpha_1~X(n-1) + \alpha_2~X(n-2) ~ + ~\cdots~ +~\alpha_p~X(n-p) + W(n) = \sum_{i=1} ^p \alpha_i ~ X(n-i) + W(n) $$

es simple determinar la función de transferencia, 

$$ H(\omega)= { 1 \over {1 - \alpha_1 ~ e^{-j\omega} - \alpha_2 ~ e^{-j2\omega} - \cdots - \alpha_p ~ e^{-jp\omega}}} ~~~|\omega|<\pi$$

Resultando la densidad espectral de potencia:

$$ S_{xx}(\omega)~ = \left| H(\omega) \right|^2 ~ S_{WW}(\omega)~=~ { \sigma_W^2 \over {\left |1 - \alpha_1 ~ e^{-j\omega} - \alpha_2 ~ e^{-j2\omega} - \cdots - \alpha_p ~ e^{-jp\omega} \right |^2}} ~~~|\omega|<\pi$$




## Aplicación: Estimación de ruido

Un ejemplo práctico, podemos suponer que se busca enviar información por un canal físico, la atmósfera por ejemplo, pero ese canal no es exclusivo, se transmiten otras señales:

<img src="img/canal.png" width="400">

Con el fin de no interferir bandas cercanas, se va a diseñar un sistema que controle el ancho de banda empleado. Para ello es necesario generar un modelo del canal y por lo tanto se propone:

- Adquirir una muestra del canal $Y=[y_0,~ y_1, ~ y_2, ~\cdots ~, y_M]$
- Generar un modelo AR-p del canal.

### Repaso: Ecuaciones de Yule - Walker

Conocida $R_{YY}(m)$ (o estimada), y considerando que para un AR-p se puede escribir:

$$ R_{YY}(m) = \alpha_1~ R_Y(m-1) + \alpha_2~R_Y(m-2) ~ + ~...~ +~\alpha_p~R_Y(m-p) = \sum_{i=1} ^p \alpha_i ~ R_{YY}(m-i) $$

se puede desarrollar el siguiente sistema de ecuaciones, empleando los coeficientes de correlación : $r_y(m)={R_{YY} (m) / \sigma_Y^2}$:


$$
 \begin{bmatrix}
   1 & r_Y(1) & r_Y(2) & \cdots & r_Y(p-1)\\
   r_Y(1) & 1 & r_Y(1) & \cdots & r_Y(p-2)\\
   r_Y(2) & r_Y(1) & 1 & \cdots & r_Y(p-3)\\
   \vdots & \vdots & \vdots & \ddots & \vdots \\
   r_Y(p-1) & r_Y(p-2) & r_Y(p-3) & \cdots & 1
   \end{bmatrix} ~ 
 \begin{bmatrix}
   \alpha_1 \\
   \alpha_2 \\
   \alpha_3 \\
   \vdots \\
   \alpha_p 
   \end{bmatrix} 
   = 
 \begin{bmatrix}
   r_Y(1) \\
   r_Y(2) \\
   r_Y(3) \\
   \vdots \\
   r_Y(p) 
   \end{bmatrix}
$$


$$\underline{R}~\underline{\alpha }= \underline{r} $$


Asumiendo que el proceso $Y(n)$ es **estacionario en sentido amplio**, se tiene que $\underline R$ es definida positiva, y por lo tanto de rango completo, entonces se obtienen los coeficientes del filtro AR-p:


$$\underline{\alpha} = \underline{R}^{-1} ~ \underline{r}$$

### Adquisición de una muestra del canal $y(n)$

```python
# Se genera artificialmente muestras de ruido.(AR-4)

# Definmos los coeficientes de un filtro, 
a = [1, -2.2137, 2.9403, -2.1697, 0.9606]
# Adquirimos las muestras (y) de ruido, generadas a partir de ruido gaussiano
y = scipy.signal.lfilter([1], a, randn(1, 1024))

# Grafiquemos qué es lo que tenemos.
fig=figure(figsize=(20,8))
ax1=fig.add_subplot(1,2,1)
xx=list(range(0,200))
plt.plot(xx,y.squeeze()[xx])
ax1.grid()
ax1.set_xlabel('muestras')
ax1.set_ylabel('amplitud')
ax1.set_title(r'Ruido adquirido Y(n)')
ax2=fig.add_subplot(1,2,2)
p = Periodogram(y.squeeze(), sampling=2)  # determinamos el periodograma
p.plot(ax=ax2)
tt=ax2.set_title(r'Periodograma $\hat Y (\omega)$' )
```

Se determinan 30 coeficientes empleando Yule-Walker

```python
AR, P, k = aryule(y[0], 30)
fig=figure(figsize=(8,6))
plt.stem(np.asarray(range(len(AR)))+1,AR)
xlabel('Coeficiente nro:')
title('Coeficientes de la estimación Yule-Walker (AR-30)')
tt=ylabel('Amplitud')
```

Se puede apreciar que la cantidad de coeficientes con amplitu relevante es mucho menor
